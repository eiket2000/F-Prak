# Lineare Regression
In dieser Aufgabe soll eine 2 dimensionale Lineare Regression implementiert werden.
## Import
Als erstes werden nötige Module importiert.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress as scipy_linregress
```

## Einlesen von Daten
Danach müssen die Daten die Leider nicht als csv-Dateien gegeben waren noch einlesen und die x und y Datensätze in Listen abspeichern

```python
data=""""0 4.910131 2.196544
1 1.500393 2.776927
2 2.946845 3.316272
3 6.102233 3.459587
4 5.168895 4.685546
5 -1.943195 0.689859
6 2.875221 2.885446
7 0.121607 1.942890
8 0.241953 2.838975
9 1.526496 3.192628
10 0.860109 2.335506
11 4.135684 3.429786
12 2.402594 2.276885
13 0.804188 1.250858
14 1.609658 2.308941
15 1.334186 2.478430
16 4.235198 3.885705
17 -0.012896 2.597321
18 1.282669 2.191137
19 -1.635239 1.358277"""
data=data.split("\n")
x=[]
y=[]
for line in data:
    line=line.split(" ")
    x.append(float(line[1]))
    y.append(float(line[2]))
x=np.array(x)
y=np.array(y)
```

# 2
## Lineare Regression
Wir wissen aus der Vorlesung, dass:

$$\hat{\beta}=(X^T X)^{-1}X^T\mathbf{y}$$

mit X:
$$X=\begin{pmatrix} 1 & x_1 \\ 1&x_2 \\ \vdots&\vdots\\1&x_N \end{pmatrix}$$
Dies ist dann der Satz an Parameteren um den LSEE zu minimieren bezüglich der Gleichung:
$$y_i=\beta_1\cdot x+\beta_0$$
zur vereinfachung werden die in Numpy verwendeten Algorithmen zur Matrixrechnung verwendet.

```python
X=np.array([[1,xi] for xi in x])
beta=np.linalg.inv(X.transpose()@X)@X.transpose()@y
print("Die Parameter sind",beta)
def func_y(x):
    return beta[1]*x+beta[0]
```

## Plotten
Zum Schluss werden die Ergebnisse mittel Matplotlib geplottet. Der Fit passt gut

```python
fig = plt.figure()
fig1 = fig.add_subplot()
fig1.set_title("Datensatz mit linearer Regeression")
fig1.set_ylabel("y")
fig1.set_xlabel("x")
fig1.plot(x,y,"o",label="Rohdaten")
fig1.plot(x,func_y(x),label="Linearer Fit")
plt.tight_layout()
plt.legend()
plt.show()
```

# 2

Nun soll das Ergebnis mit der in scipy integrierten Regression verglichen werden. Wenn man diese plottet sieht man, dass diese sehr ähnlich ist.

```python
scipy_para=[scipy_linregress(x,y)[1],scipy_linregress(x,y)[0]]

fig = plt.figure()
fig1 = fig.add_subplot()
fig1.set_title("Datensatz mit linearer Regeression")
fig1.set_ylabel("y")
fig1.set_xlabel("x")
fig1.plot(x,y,"o",label="Rohdaten")
fig1.plot(x,scipy_para[1]*x+scipy_para[0],label="Linearer Fit")
plt.tight_layout()
plt.legend()
plt.show()
```

Ein kurzer Vergleich zeigt, dass die Parameter fast bis auf Maschinengenauigkeit identisch sind, daher sind sie wahrscheinlich gleich implementiert.

```python
print(beta-scipy_para)
```
