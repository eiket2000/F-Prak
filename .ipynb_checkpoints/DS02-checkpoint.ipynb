# 4 Time Series
Importiere genutzte Module

```python
import numpy as np
import matplotlib.pyplot as plt
```

Füge Funktionen hinzu um die benötigten Normen auszurechnen

```python
def L1(A,B):
    return np.sum(np.abs(A-B))
def Linfty(A,B):
    return np.max(np.abs(A-B))

```

## 4.1
Berechne nun die beiden Normen zwischen den beiden Zeitreihen

```python
A=np.array([-1.66,0.3,-0.08,.1,-1.17,-0.05,0.84,-0.66,0.42,-0.99])
B=np.array([0.29,0.89,0.82,0.97,0.53,0.83,1.06,0.67,0.86,0.51])

print("L1 Norm:",L1(A,B),"L infinity Norm:",Linfty(A,B))
```

## 4.2
Bereinige die beiden Datensatz um den Offset und berechne dann wieder die Normen.

```python
def offset(A):
    return A-np.mean(A)

A_offset=offset(A)
B_offset=offset(B)

print("L1 Norm:",L1(A_offset,B_offset),"L infinity Norm:",Linfty(A_offset,B_offset))
```

Man sieht, dass beide Normen bedeutend kleiner geworden sind.
## 4.3
Führe nun auf diesen um den offset bereinigten Daten auch noch eine Amplitudenskalierung durch.

```python
def amplitude(A):
    return (A-np.mean(A))/np.std(A)

A_amp=amplitude(A_offset)
B_amp=amplitude(B_offset)
print("L1 Norm:",L1(A_amp,B_amp),"L infinity Norm:",Linfty(A_amp,B_amp))
```

Man sieht, dass auch hier beide Normen sich ungefähr halbiert haben.

## Plots
zur beseeren Visualisierung wurden auch alle 3 Versionen der beiden Zeitreihen geplottet.

```python
fig = plt.figure()
fig1 = fig.add_subplot()
fig1.set_title("Rohdaten")
fig1.set_ylabel("y")
fig1.set_xlabel("x")
fig1.plot(A,label="A")
fig1.plot(B,label="B")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
fig = plt.figure()
fig1 = fig.add_subplot()
fig1.set_title("Daten mit Offset skalierung")
fig1.set_ylabel("y")
fig1.set_xlabel("x")
fig1.plot(A_offset,label="A")
fig1.plot(B_offset,label="B")
plt.tight_layout()
plt.legend()
plt.show()


```

```python
plt.show()
fig = plt.figure()
fig1 = fig.add_subplot()
fig1.set_title("Daten mit Offset und Amplituden skalierungen")
fig1.set_ylabel("y")
fig1.set_xlabel("x")
fig1.plot(A_amp, label="A")
fig1.plot(B_amp, label="B")
plt.tight_layout()
plt.legend()
plt.show()
```
